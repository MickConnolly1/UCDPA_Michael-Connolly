# Will Further Gun Control Reduce The Possibilities Of Mass Shootings

```python
# extracting the required columns with loc function for "gender", "age", "region", "will_further_gun_control_reduce_the_possibilities_of_mass_shootings"
reduce_killings = gun_control.loc[:, ["gender", "age", "will_further_gun_control_reduce_the_possibilities_of_mass_shootings"]]
print(reduce_killings.head())        # printing first 5 rows of results
```

```python
# I will create a countplot with a pie chart in the center this will hold my values yes and no
plt.figure(figsize=(12,8))
sns.set_style("whitegrid")
ax = sns.countplot(x="will_further_gun_control_reduce_the_possibilities_of_mass_shootings", data=reduce_killings, palette="Set2")
ax.set_title("Opinions on Further Gun Control for Reducing Mass Shootings")

# Adding the pie chart in the center
total_responses = len(reduce_killings)
pos_responses = (reduce_killings["will_further_gun_control_reduce_the_possibilities_of_mass_shootings"] == "Yes").sum()
neg_responses = (reduce_killings["will_further_gun_control_reduce_the_possibilities_of_mass_shootings"] == "No").sum()
plt.axes([0.45, 0.45, 0.1, 0.1])
plt.pie([pos_responses, neg_responses], labels=["Yes", "No"], colors=["springgreen", "orange"], autopct="%1.1f%%")
plt.axis("equal")
plt.title("{:.1f}% Yes\n{:.1f}% No".format(pos_responses/total_responses*100, neg_responses/total_responses*100))

# Show the chart
plt.show()
```

# Adding a Second CSV Dataframe to Merge with Gun_Control Dataframe

```python
# source of secondary data. https://www.kaggle.com/datasets/jameslko/gun-violence-data?resource=download.
secondary_data = pd.read_csv(r"C:\Users\micha\Downloads\gun-violence-data_2.csv")
secondary_data.info()
secondary_data.describe()
secondary_data.head()
secondary_data.shape
secondary_data.isnull()
secondary_data.dropna()

# extracting the columns i want to merge with dataframe 1 gun_control with "state" being the common value in both sets
gun_control_2 = secondary_data.loc[:, ["state", "city_or_county", "date", "n_killed", "n_injured"]] # extracting the columns i want to use, using the Loc funtion
```

```python
mask_1 = gun_control_2.duplicated()
gun_control_2 = gun_control_2.drop(mask_1.index[mask_1])
print(gun_control_2.head())     
# printing the new data set minus duplicated values.
```

```python
# merging my 2 data frames and using outer join as dataframe 1 has values in "state" as CSV values and dataframe 2 has single values
merged_data = pd.merge(gun_control, gun_control_2, on="state",how="outer")  # using outer join on "state" its common value
merged_data.fillna("Not Available",inplace=True) # replacing NaN values with "Not Available"
merged_data.info()
```

```python
#  creating a new DataFrame by grouping the "merged_data" by "date" and "state" columns  and calculating the sum of the "n killed" column
#  a new "year" column is created by splitting the "date" column using "/" as delimiter for the year only and dropping the rest using  axis 1 method 
#  analyzing the number of people killed in different states on different dates
#  years are created in a new column based on "date" info

dataframe_1 = merged_data.groupby(["date", "state"])[["n_killed"]].sum().reset_index()
dataframe_1["year"] = dataframe_1["date"].str.split("/").str[-1]
dataframe_1 = dataframe_1.drop("date", axis=1)
print(dataframe_1.head())                    
```

```python
# create a pivot table where the rows are the years and the columns the states
pivot_table = dataframe_1.pivot_table(index=["year"], columns=["state"], values="n_killed", aggfunc="sum")

# create the bar chart
ax = pivot_table.plot(kind="bar", stacked=False, figsize=(18,8),width=2.5)

# adding names for the labels and the title of the chart
ax.set_xlabel("year")
ax.set_ylabel("Total Killed")
ax.set_title("Total Number of People Killed per State for each Year")

# show the chart
plt.show()
```

```python
# using "secondary_data" and selecting the four specific columns i want  ("state", "date", "n_killed", "n_injured") using the `loc` function
# grouping the data by state using the "groupby" function and calculating the sum of the "n_killed" and "n_injured" fields for each state
# analyse gun control and its effects on fatalities and injuries in different states.

gun_control_2 = secondary_data.loc[:, ["state", "date", "n_killed", "n_injured"]]
grouped_data = gun_control_2.groupby("state").sum()
print(grouped_data.head())        # printing first 5 rows
```

```python
# Create a scatter plot of n_killed vs n_injured for each state, with labels for each point
fig, ax = plt.subplots(figsize=(15,18))
ax.scatter(grouped_data["n_killed"], grouped_data["n_injured"])

for i, txt in enumerate(grouped_data.index):
    ax.annotate(txt, (grouped_data["n_killed"][i], grouped_data["n_injured"][i]))

plt.xlabel("Number of people killed",fontsize=18)
plt.ylabel("Number of people injured",fontsize=18)
plt.title("Gun violence in the US by state between 2013-2018",fontsize=24)
plt.show()

# data suggests that "illinois, california,florida,texas" have the higest number of gun violence between 2013 and 2018 (years are shown in previous graph)
```

```python
# Get the top 3 results by sorting by the numbers killed ("n_killed")
top_3 = grouped_data.sort_values('n_killed', ascending=False)[:3]

# Creating a numpy array of the top 3 states and their numbers killed ("n_killed") values
top_3_array = np.array([(top_3.index.values, top_3['n_killed'].values)])
display = f"The top 3 states with the highest killings and injuries are {', '.join(top_3.index.values)}." # storing "F" string to a variable called display
# Print the message with the top 3 countries of numbers killed
print(display)
```

```python
gun_control_2 = secondary_data.loc[:, ["state","incident_characteristics","participant_type"]]
grouped_data = gun_control_2.groupby("state").sum()
print(grouped_data.head())       # printing first 5 rows
```

```python
# Grouping data by the "state"
grouped_by_state = gun_control_2.groupby("state")
grouped_by_state.fillna("Not Available",inplace=True)

# Grouping data by the "incident_characteristics" 
grouped_by_incident = gun_control_2.groupby("incident_characteristics")
grouped_by_incident.fillna("Not Available",inplace=True)

# Grouping data by the "participant_type" 
grouped_by_participant = gun_control_2.groupby("participant_type")
grouped_by_participant.fillna("Not Available",inplace=True)

# Print the first 5 rows of each group
print(grouped_by_state.head())
print(grouped_by_incident.head())
print(grouped_by_participant.head())
```

```python
# creating a new dataframe and extractin "Gang Involvement" from "incident_characteristics" column as they are 2 many choices to display in graph.
# str.contains value can be changed to reference other types of gun crimes in the "incident_characteristics"
# created a new dataframe called "gang_involvement
gang_involvement = gun_control_2[gun_control_2["incident_characteristics"].str.contains("Gang involvement", na=False)]
gang_involvement = gang_involvement.fillna(value="Not Available")

# Grouping data by the "state"
grouped_by_state = gang_involvement.groupby("state")

# Print the first 5 rows of each group
print(grouped_by_state.head())
```

```python
# graph represents gun violence by "gang involvement" per state.

sns.set_style("whitegrid", {"font.family": "sans-serif", "font.serif": ["Times", "Arial"]})

sns.set(rc={"figure.figsize":(12,6)})
ax = sns.countplot(y="state", data=gang_involvement, palette="Set2")
ax.set_title("Number of Incidents with Gang Involvement by State", fontsize=16, fontweight='bold')
ax.set_xlabel("Count", fontsize=18)
ax.set_ylabel("State", fontsize=18)
ax.tick_params(labelsize=10)
plt.tight_layout()

# Showing the plot
plt.show()
```
